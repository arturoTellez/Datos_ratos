```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('dark_background')
```

edad: edad del beneficiario principal

sexo: sexo del contratista de seguros, mujer, hombre

bmi: índice de masa corporal, que proporciona una comprensión del cuerpo, pesos relativamente altos o bajos en relación con la altura,
índice objetivo de peso corporal (kg / m ^ 2) utilizando la relación entre la altura y el peso, idealmente 18,5 a 24,9

hijos: Número de hijos cubiertos por el seguro médico / Número de dependientes

fumador: fumar

región: el área residencial del beneficiario en los EE. UU., noreste, sureste, suroeste, noroeste.

Cargos: Costos médicos individuales facturados por el seguro médico.

```python
df = pd.read_csv("data/insurance.csv")
print(df.info())
df.head()
```

```python
variable = "charges"
plt.figure(figsize = (11, 7))
df[variable].hist()
plt.title("Histograma de variable {}".format(variable), fontsize = 20)
plt.xlabel(variable, fontsize = 20)
plt.ylabel("Número de registros", fontsize = 20)
print(df[variable].describe(np.linspace(0,1,11)))
plt.show()
```

```python
v_num = ["age", "bmi", "children", "charges"]
v_cat = ["sex", "smoker", "region"]
```

```python
for variable in v_num:
    print(variable)
    plt.figure(figsize = (11, 7))
    df[variable].hist()
    plt.title("Histograma de variable {}".format(variable), fontsize = 20)
    plt.xlabel(variable, fontsize = 20)
    plt.ylabel("Número de registros", fontsize = 20)
    print(df[variable].describe(np.linspace(0,1,11)))
    plt.show()
```

```python
for variable in v_cat:
    print(variable)
    plt.figure(figsize = (11, 7))
    df[variable].value_counts().plot.bar()
    plt.title("Histograma de variable {}".format(variable), fontsize = 20)
    plt.xlabel(variable, fontsize = 20)
    plt.ylabel("Número de registros", fontsize = 20)
    print(df[variable].value_counts())
    plt.show()
```

```python
df["altura"] = np.random.normal(1.59, (.08)**(1/2), size = df.shape[0])
df["peso"] = df.bmi * df.altura
df["altura"].hist()
```

```python
variable = "peso"
plt.figure(figsize = (11, 7))
df[variable].hist()
plt.title("Histograma de variable {}".format(variable), fontsize = 20)
plt.xlabel(variable, fontsize = 20)
plt.ylabel("Número de registros", fontsize = 20)
print(df[variable].describe(np.linspace(0,1,11)))
plt.show()
```

```python
variable = "altura"
plt.figure(figsize = (11, 7))
df[variable].hist()
plt.title("Histograma de variable {}".format(variable), fontsize = 20)
plt.xlabel(variable, fontsize = 20)
plt.ylabel("Número de registros", fontsize = 20)
print(df[variable].describe(np.linspace(0,1,11)))
plt.show()
```

```python
df[["altura", "peso"]].plot.scatter(*["altura", "peso"])
```

```python
df["fuera_rango"] = np.where((df.bmi >= 18.5) & (df.bmi <= 24.9), 1,  0)
```

```python
plt.figure(figsize = (11, 7))
df[["altura", "peso", "fuera_rango"]].plot.scatter(*["altura", "peso"], c  = "fuera_rango", 
                                                   colormap='viridis', figsize = (11, 7))
plt.title("Gráfica de peso vs altura", fontsize = 20)
plt.xlabel("Altura", fontsize = 20)
plt.ylabel("Peso", fontsize = 20)
plt.show()
```
