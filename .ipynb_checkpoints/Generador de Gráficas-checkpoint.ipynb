```python
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
plt.style.use('dark_background') # https://matplotlib.org/3.1.0/gallery/style_sheets/style_sheets_reference.html
```

```python
datos = np.random.normal(30, 5, size = 10000)
datos = np.append(datos, 100)
```

```python
np.random.shuffle(datos)
```

```python
plt.figure(figsize = (11, 7))
plt.hist(datos)
plt.title("Temperatura Ambiental promedio diaria", fontsize = 20)
plt.xlabel("Grados", fontsize = 20)
plt.ylabel("Número de días", fontsize = 20)
```

```python
base = datetime.datetime.today()
date_list = [base - datetime.timedelta(days=x) for x in range(10001)]
date_list
```

```python
len(date_list) == len(datos)

```

```python
plt.figure(figsize = (11, 7))
plt.scatter(date_list, datos)
plt.title("Temperatura Ambiental promedio diaria", fontsize = 20)
plt.xlabel("Año", fontsize = 20)
plt.ylabel("°C", fontsize = 20)
```
